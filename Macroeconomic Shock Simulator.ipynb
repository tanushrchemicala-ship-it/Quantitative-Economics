# Macroeconomic Shock Simulator

```python

#Economy Initialisation
import random as rn
economy = np.zeros((120 , 4))
mpc = 0.8 + rn.uniform(-0.05 , 0.05 )
g = 500
base_i = 300

shocks = np.random.normal(0 , 50 , 120) # Generating yearly random shocks to simulate real life macroeconomic uncertainties

interest_rates = np.zeros(120)
interest_rates[0:60] = 0.03
interest_rates[60:120] = 0.08

c = 1000
i = base_i - 1000*interest_rates[0] 
y = c + i + g

economy[0 , 0] = y
economy[0 , 1] = c
economy[0 , 2] = i
economy[0 , 3] = g

print(f"Month 1: {economy[0 , :]}")

#Running Simulation

for x in range(1 , 120):
    economy[x , 3] = g
    economy[x , 2] = base_i - 1000*interest_rates[x]
    economy[x , 1] = mpc * economy[x - 1 , 0]
    economy[x , 0] = economy[x , 1] + economy[x , 2] + economy[x , 3] + shocks[x]

economy[45 , 0]*=0.9 #Simulating a Black-Swan event

print(f"Month 60 : {economy[59 , :]}")
print(f"Month 120: {economy[-1 , :]}")

#Data Insights

lost_output = economy[59 , 0] - economy[-1 , 0]
print(f"The total output lost due to the midway hike in interest rates was Rs.{lost_output}")


```
