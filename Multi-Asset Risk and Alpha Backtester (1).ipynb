# MULTI-ASSET RISK & ALPHA BACKTESTER

```python
#Market Simulation and Data Integrity 
import numpy as np

market_data = np.random.normal(0.001 , 0.02 , size=(252 , 5))

random_1 = np.random.randint(0 , 252 , 10)
random_2 = np.random.randint(0 , 5 , 10)

market_data[random_1 , random_2] = np.nan
mean_value = np.nanmean(market_data)

mask = np.isnan(market_data)
market_data[mask] = mean_value #Simulating and repairing missing market data to ensure pipeline robustness.


#Portfolio Construction

weights = np.random.rand(5)
weights = weights/np.sum(weights)

port_returns = market_data @ weights

# The Alpha Strategy

trade_signals = np.where(port_returns < -0.01 , 1.5 , np.where(port_returns > 0.01 , 0.5 , 1)) #Overweighting (1.5x) on dips and underweighting (0.5x) on spikes.

strat_returns = port_returns * trade_signals

# Risk Audit

extreme_loss = strat_returns < -0.02
strat_returns[extreme_loss] = -0.02 #Setting a stop loss floor of -2%


#Performance Analysis

strat_growth = strat_returns + 1
strat_wealth = np.round(10000 * np.cumprod(strat_growth) , 2)
normal_wealth = np.round(10000 *np.cumprod(1+port_returns) , 2)
actual_volatility = np.std(strat_returns)


print(f"Final Strategized Portfolio Wealth: Rs.{strat_wealth[-1]}")
print(f"Final Normal Portfolio Wealth: Rs.{normal_wealth[-1]}")
```
