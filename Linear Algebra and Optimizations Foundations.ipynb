```python
#Question 18

A = np.array([
    [2.0, 0.8, 0.5],
    [0.8, 1.5, 0.3],
    [0.5, 0.3, 1.0]
])

eigenvalues , eigenvectors = np.linalg.eig(A)

B = np.array(eigenvectors)
B_inverse = np.linalg.inv(B)
diagonal_matrix = np.diag(eigenvalues)

A_reconstructed = np.round(B @ diagonal_matrix @ B_inverse , 2)

print(A_reconstructed)
print("Eigen Decomposition Completed")
print(f"Eigen Values: {eigenvalues}")
```

```python
#Question 27

def numerical_gradient(x , y , h=10**-5):
    return (function(x+h , y) - function(x , y)) / h

print(numerical_gradient(1 , 2))
```
